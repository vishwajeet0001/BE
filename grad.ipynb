```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun(x):
    return (x+3)**2
def grad(x):
    return 2*(x+3)

x=2
lr=0.1
epchos=100

x_v=[x]
y_v=[fun(x)]

for i in range(epchos):
    x=x-lr*grad(x)
    x_v.append(x)
    y_v.append(fun(x))

print("local min",x)
print("value of function",fun(x))
```

```python
x_range=np.linspace(-10,5,100)
y_range=fun(x_range)

plt.plot(x_range,y_range,label='y=(x+3)**2',color='green')
plt.scatter(x_v,y_v,marker="*",color="red",label="gradient")
plt.legend()
plt.grid(True)
```
